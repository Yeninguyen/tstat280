Write a single SQL query to analyze when customers place orders.

- Join `customers` and `orders` using a LEFT JOIN
- Use the relationship between primary and foreign keys (customer_id)

Filter the data so that:
- Customers are in states 'WA', 'OR', or 'CA'
- AND their first name starts with 'J'
- OR they are in the city 'Portland'

For each customer, calculate:
- first_order_date (MIN of order_date)
- last_order_date (MAX of order_date)

Only include customers whose most recent order was after '2017-01-01'.

Sort the results by last_order_date in descending order.

Limit the results to the top 5 customers.

Include the following columns:
- customer_id
- first_name
- last_name
- first_order_date
- last_order_date

```python
import sqlite3

conn = sqlite3.connect("../data/bike_store.db")
curr = conn.cursor()

curr.execute("""
    SELECT customers.customer_id, customers.first_name, customers.last_name, 
    MIN(orders.order_date) AS first_order_date, MAX(orders.order_date) AS last_order_date
    FROM customers
    LEFT JOIN orders ON customers.customer_id = orders.customer_id
    WHERE state IN ('WA', 'OR', 'CA')
    AND (customers.first_name LIKE 'J%' OR city = 'Portland')
    GROUP BY customers.customer_id
    HAVING orders.order_date >'20170101'
    ORDER BY last_order_date DESC
    LIMIT 5
    
""").fetchall()
```
