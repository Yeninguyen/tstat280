### Practice Quiz

#### Part 1

Create a list called `numbers` with the values `[1, 2, 3, 4, 5]`

```python
numbers = [1, 2, 3, 4, 5]
print(numbers)
```

#### Part 2

Use a **list comprehension** to create a new list called `squared` that contains each number squared

```python
numbers = [1, 2, 3, 4, 5]
squared = [i**2 for i in numbers]
print(squared)
```

#### Part 3

Create a dictionary called `results` with:
   - `"original"` → the `numbers` list  
   - `"squared"` → the `squared` list

Then print the keys of this dictionary

```python
numbers = [1, 2, 3, 4, 5]
squared = [i**2 for i in numbers]
results = {
    "original" : numbers,
    "squared" : squared
}
for key in results:
    print(key)
```

#### Part 4

Use **indexing/slicing** to:
   - Print the first three elements of `squared`  
   - Print the last element of `numbers`  

```python
numbers = [1, 2, 3, 4, 5]
squared = [i**2 for i in numbers]
print(squared[0:3])
print(numbers[4:])
```

#### Part 5

Write a **for loop** that:
   - Iterates through `squared`  
   - Prints `"Greater than 10"` if the number is greater than 10, otherwise prints `"10 or less"`  

```python
numbers = [1, 2, 3, 4, 5]
squared = [i**2 for i in numbers]
for i in squared:
    if i > 10:
        print("Greate than 10")
    else:
        print("10 or less")
    
```

#### Part 6

Import the `math` library and define a function called `divide_and_round` that:
   - Takes two numbers as input  
   - Divides the second number by the first number  
   - Returns the result rounded **down** using `math.floor()`   

```python
import math
def devide_and_round(a,b):
    return math.floor(b/a)
print(devide_and_round(3,10))
```

#### Part 7

Call your function using the last two elements of `numbers` and print the result  

```python
import math
numbers = [1, 2, 3, 4, 5]
def devide_and_round(a,b):
    return math.floor(b/a)
print(devide_and_round(numbers[3],numbers[4]))
```
